```python
import pandas as pd
```

```python
data=pd.read_csv(r"D:\machine learning\apriory algorithm\GroceryStoreDataSet.csv",names=['Product'])
```

```python
data
```

```python
data.columns
```

```python
data.values
```

```python
data=list(data['Product'].apply(lambda x:x.split(',')))
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_data=te.fit(data).transform(data)
df=pd.DataFrame(te_data, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df1=apriori(df, min_support=0.01, use_colnames=True)
df1
```

```python
df1['length']=df1['itemsets'].apply(lambda x:len(x))
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud

plt.rcParams['figure.figsize'] = (12, 5)
wordcloud = WordCloud(background_color = 'white', width = 1200,  height = 1200, max_words = 121).generate(str(df1['itemsets']))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Items',fontsize = 20)
plt.show()
```
