```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("D:\machine learning\knn\Iris.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
df=data.drop('Species', axis=1)
df.head()
```

```python
df.isnull().any()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x=df['SepalLengthCm'][:50],y=df['SepalWidthCm'][:50],data=df)
plt.title("sepal length and sepal width")
```

```python
# the histogram of the data
plt.hist(df, 10, density=False, facecolor='green')
#density = false means our width of bar will be same
plt.xlabel(df['PetalLengthCm'].iloc[0])
plt.ylabel('PetalWidthCm')
plt.title('Histogram Example')

plt.grid(True)

plt.show()
```

```python
df.columns
```

```python
features=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
select_df=df[features]
```

```python
select_df.head()
```

```python
X = StandardScaler().fit_transform(select_df)
X
```

```python

kmeans = KMeans()
model = kmeans.fit(X)
```

```python
centers = model.cluster_centers_
centers
```
