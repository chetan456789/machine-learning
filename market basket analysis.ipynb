```python
df = list(data["Item"].apply(lambda x:x.split(',')))
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(df).transform(df)
dataset = pd.DataFrame(te_data,columns=te.columns_)
dataset
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df3 = apriori(dataset, min_support=0.01,use_colnames=True)
df3
```

```python
df3.sort_values(by="support",ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud

plt.rcParams['figure.figsize'] = (12, 5)
wordcloud = WordCloud(background_color = 'white', width = 1200,  height = 1200, max_words = 121).generate(str(df3['itemsets']))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Items',fontsize = 20)
plt.show()
```
