```python
#Checking hypothesis: "More the number of colonies, higher the production value"
sns.lmplot(x="X5 latitude", y="X6 longitude", data=df)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop(['Y house price of unit area'],axis=1), 
                                                    df['Y house price of unit area'], test_size=0.1, 
                                                    random_state=10)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
accuracy=model.score(X_test,y_test)
print(accuracy*100,'%')
```
